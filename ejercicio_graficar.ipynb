```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3
import pandas as pd
import numpy as np


# Datos
data = {
    'Sistema de Agrupación': ['Bit', 'Crumb', 'Nyble', 'Byte', 'Octet', 'Word', 'Double Word (DWORD)', 'Quad Word (QWORD)', 'Double Byte', 'Double Quad Word', 'Yottabyte', 'Brontobyte'],
    'Tamaño en Bits': [1, 2, 4, 8, 8, 16, 32, 64, 16, 128, 2**80 * 8, 2**90 * 8],
    'Descripción': [
        'Unidad más pequeña de información.',
        'Agrupación de 2 bits.',
        'Similar al nibble, pero con diferente ortografía.',
        'Unidad básica de almacenamiento.',
        'Similar al byte, usado en redes.',
        'Usado en arquitecturas de 16 bits.',
        'Usado en arquitecturas de 32 bits.',
        'Usado en arquitecturas de 64 bits.',
        'Agrupación de 2 bytes.',
        'Agrupación de 2 QWORDs.',
        'Una de las unidades más grandes en almacenamiento, aunque rara vez se usa en la práctica.',
        'Concepto teórico de almacenamiento aún mayor que el yottabyte, raramente utilizado.'
    ],  'arquitectura ' : ['-', '-', '-', '8 bits', '8 bits', '16 bits', '32 bits', '64 bits', '16 bits', '128 bits', '-', '-']
}

# Crear DataFrame
df = pd.DataFrame(data)

# Ordenar por Tamaño en Bits
df = df.sort_values(by='Tamaño en Bits')

# Mostrar tabla en consola
print(df)


# Datos
sistemas = ['Bit', 'Crumb', 'Nyble', 'Byte', 'Octet', 'Word', 'Double Word (DWORD)', 'Quad Word (QWORD)', 'Double Byte', 'Double Quad Word', 'Yottabyte', 'Brontobyte']

tamanos_bits = [1, 2, 4, 8, 8, 16, 32, 64, 16, 128, 2**80 * 8, 2**90 * 8]

arquitecturas_1 = [0, 0, 0, 8 , 8 , 16 , 32 , 64 , 16, 128 , 0, 0]
arquitecturas = ['x86', 'x64', 'ARM', 'RISC-V']
tamanos_architecturas = {
    'x86': {'Bit': 1, 'Byte': 8, 'Word': 16, 'Double Word (DWORD)': 32, 'Quad Word (QWORD)': 64},
    'x64': {'Bit': 1, 'Byte': 8, 'Word': 16, 'Double Word (DWORD)': 32, 'Quad Word (QWORD)': 64},
    'ARM': {'Bit': 1, 'Byte': 8, 'Word': 32, 'Double Word (DWORD)': 64, 'Quad Word (QWORD)': 128},
    'RISC-V': {'Bit': 1, 'Byte': 8, 'Word': 32, 'Double Word (DWORD)': 64, 'Quad Word (QWORD)': 128}
}



# Crear el diagrama de barras para tamaños en bits
plt.figure(figsize=(10, 6))
plt.barh(sistemas, tamanos_bits, color='skyblue')
plt.xlabel('Tamaño en Bits')
plt.title('Comparación de Sistemas por Tamaño en Bits')
plt.xscale('log')  # Escala logarítmica para mejor visualización
plt.grid(True, which="both", ls="--")
plt.show()

"""# Verificar que las listas tengan la misma longitud
if len(sistemas) != len(arquitecturas_1):
    raise ValueError("Las listas 'sistemas' y 'arquitecturas_1' deben tener la misma longitud")
"""

# Crear el diagrama de barras para arquitecturas
plt.figure(figsize=(10, 6))
plt.barh(sistemas, arquitecturas_1, color='skyblue')
plt.xlabel('Agrupado en Bits')
plt.title('Comparación de Sistemas por Agrupacion en Bits')
plt.xscale('log')  # Escala logarítmica para mejor visualización
plt.grid(True, which="both", ls="--")
plt.show()

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(14, 8))
# Create a figure and axis object
fig, ax = plt.subplots()

# Configurar el índice de barras y el ancho de las barras
bar_width = 0.2
index = np.arange(len(sistemas))

# Crear barras para cada arquitectura
for i, arquitectura in enumerate(arquitecturas):
    tamanos = [tamanos_architecturas[arquitectura].get(sistema, 0) for sistema in sistemas]
    ax.bar(index + i * bar_width, tamanos, bar_width, label=arquitectura)

# Añadir etiquetas y título
ax.set_xlabel('Sistemas de Agrupación')
ax.set_ylabel('Tamaño (bits)')
ax.set_title('Comparación de Sistemas de Agrupación de Bits en Diferentes Arquitecturas')
ax.set_xticks(index + bar_width * (len(arquitecturas) - 1) / 2)
ax.set_xticklabels(sistemas, rotation=45, ha='right')
ax.legend()

# Mostrar gráfico
plt.tight_layout()
plt.show()


"""
# Diagrama de Venn
# Para este ejemplo, usaremos un diagrama de Venn de 3 conjuntos para ilustrar la comparación
venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels=('Bits Pequeños', 'Bits Medianos', 'Bits Grandes'))
plt.title('Comparación de Tamaños en Bits')
plt.show()}
"""

```
