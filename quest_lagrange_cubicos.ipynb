```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
from numpy.polynomial.polynomial import Polynomial
```

```python
# Datos proporcionados en el enunciado
x_points = [0, 1, 2, 3, 4, 5, 6]
y_points = [3, 5, 6, 5, -2, 2, 13]

# Interpolación de Lagrange
lagrange_poly = Polynomial.fit(x_points, y_points, deg=6).convert()

# Trazadores cúbicos
spline = CubicSpline(x_points, y_points)

# Puntos para graficar
x_dense = np.linspace(min(x_points), max(x_points), 200)
y_lagrange = lagrange_poly(x_dense)
y_spline = spline(x_dense)

# Evaluar en x = 3.25
x_eval = 3.25
y_lagrange_eval = lagrange_poly(x_eval)
y_spline_eval = spline(x_eval)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(x_dense, y_lagrange, label='Polinomio de Lagrange', linestyle='--', color='orange')
plt.plot(x_dense, y_spline, label='Trazadores Cúbicos', color='blue')
plt.scatter(x_points, y_points, color='red', label='Puntos Originales')
plt.scatter([x_eval], [y_lagrange_eval], color='green', label=f'Lagrange en x={x_eval:.2f}: {y_lagrange_eval:.2f}', marker='x', s=100)
plt.scatter([x_eval], [y_spline_eval], color='purple', label=f'Trazadores cubicos en x={x_eval:.2f}: {y_spline_eval:.2f}', marker='o', s=100)
plt.legend()
plt.title("Comparación de Interpolación de Lagrange y Trazadores Cúbicos")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

print(f"Valor estimado por Lagrange en x = 3.25: {y_lagrange_eval}")
print(f"Valor estimado por Spline en x = 3.25: {y_spline_eval}")
```
